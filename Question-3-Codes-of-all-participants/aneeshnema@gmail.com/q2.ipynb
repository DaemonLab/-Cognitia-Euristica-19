```python
import pandas as pd
import numpy as np
import seaborn as sns
import xgboost as xgb
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
```

```python
train_df = pd.read_csv('data_train.csv')
train_df.head()
```

```python
sns.pairplot(train_df)
```

```python
sns.heatmap(train_df[train_df.columns[0:-1]].corr())
```

```python
X_train, X_test, y_train, y_test = train_test_split(train_df[train_df.columns[0:-1]], train_df['label'], test_size=0.2,random_state=0)
```

```python
xgb_model = xgb.XGBClassifier(learning_rate=1,max_depth=3,random_state=0)
```

```python
xgb_model.fit(X_train,y_train)
```

```python
xgb_model.score(X_test,y_test)
```
